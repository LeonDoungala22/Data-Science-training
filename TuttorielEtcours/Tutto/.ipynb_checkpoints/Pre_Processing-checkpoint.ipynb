Le prétraitement des données, une composante de la préparation des données, décrit tout type de traitement effectué sur des données brutes pour les préparer à une autre procédure de traitement des données. Il s'agit traditionnellement d'une étape préliminaire importante pour le processus d'exploration de données.

Qu'est-ce que le prétraitement des données et pourquoi est-il important ?
Le prétraitement des données est le processus de transformation des données brutes en un format utile et compréhensible. Les données brutes ou du monde réel ont généralement un formatage incohérent, des erreurs humaines et peuvent également être incomplètes. Le prétraitement des données résout ces problèmes et rend les ensembles de données plus complets et plus efficaces pour l'analyse des données.

Parmis les operations de preprocessing les plus important on retrouve :
    
- **L'Encodage** : Convertir les variables qualitative en valeurs numerique quantitative ( expl : **[chien , chat , chien , chat ] --> [0,1,0,1]** ) 
- **La normalisation** : Mettre sur une meme echelle toutes variables quantitatives ( expl : *[2,10,4,6] --> [0,1,0.25,0.5]*
- **L'imputation** : Qui permet de remplacer les donnée manquantes en certaines valeures statistiques ( exple : **[2,1,3,'nan'] --> [2,1,3,2] )**
- **Sélection**: Qui utilise les test statistique comme le test de Chi-E 
- **L'extraction** : Permet de generer de nouvelles variables à partir d'informations cachées dans le dataset

Les operations les plus utilisées sont les deux premieres .
