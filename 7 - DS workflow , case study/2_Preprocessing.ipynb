# STEP 2: PRE-PROSSESSING

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

```

```python
data = pd.read_csv('Dataset/covid.csv', index_col=0, encoding = "ISO-8859-1")
data.head()
```

## PRE-PROCESSING

```python
df = data.copy()
df.head()
```

### Creation of sub-assemblies (following the DATA_ANALISIS module)

```python
missing_rate = df.isna().sum()/df.shape[0]
```

```python
#Split dataset into two sub-assemblies 

blood_columns = list(df.columns[(missing_rate < 0.9) & (missing_rate >0.88)])
viral_columns = list(df.columns[(missing_rate < 0.80) & (missing_rate > 0.75)])
```

```python
#Check de Data Analisis step to understand why

key_columns = ['Patient age quantile', 'SARS-Cov-2 exam result']
```

```python
df = df[key_columns + blood_columns + viral_columns]
df.head()
```

### Train Test - Cleaning - Encoding

```python
from sklearn.model_selection import train_test_split
```

```python
#spilt Dataset 

trainset, testset = train_test_split(df, test_size=0.2, random_state=0)
```

```python
#Train set

trainset['SARS-Cov-2 exam result'].value_counts()
```

```python
#Test set

testset['SARS-Cov-2 exam result'].value_counts()
```

```python
#Data transformation 

def encodage(df):
    code = {'negative':0,
            'positive':1,
            'not_detected':0,
            'detected':1}
    
    for col in df.select_dtypes('object').columns:
        df.loc[:,col] = df[col].map(code)
        
    return df
```

```python
#main preprocessiong functions (basic preprocessing )

def feature_engineering(df):
    df['is sick'] = df[viral_columns].sum(axis=1) >= 1
    df = df.drop(viral_columns, axis=1)
    return df

def imputation(df):
    df = df.dropna(axis=0)
    return  df


def preprocessing(df):
    
    df = encodage(df)
    df = feature_engineering(df)
    df = imputation(df)
    
    X = df.drop('SARS-Cov-2 exam result', axis=1)
    y = df['SARS-Cov-2 exam result']
    
    print(y.value_counts())
    
    return X, y
```

```python
#Retreve X_train and y_train from preprossing 

X_train, y_train = preprocessing(trainset)
```

```python
#Retreve X_test and y_test from preprossing 

X_test, y_test = preprocessing(testset)
```

# Modelling(Basic)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import PolynomialFeatures
from sklearn.decomposition import PCA
```

```python
model_1 = RandomForestClassifier(random_state=0 ,n_estimators=25)
```

```python
#Pipeline

model_2 = make_pipeline(PolynomialFeatures(2), SelectKBest(f_classif, k=10),
                      RandomForestClassifier(random_state=0))
```

### Evaluation process

```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import learning_curve
```

```python
def evaluation(model):
    
    model.fit(X_train, y_train)
    ypred = model.predict(X_test)
    
    print(confusion_matrix(y_test, ypred))
    print(classification_report(y_test, ypred))
    
    N, train_score, val_score = learning_curve(model, X_train, y_train,
                                              cv=4, scoring='f1',
                                               train_sizes=np.linspace(0.1, 1, 10))
    
    
    plt.figure(figsize=(12, 8))
    plt.plot(N, train_score.mean(axis=1), label='train score')
    plt.plot(N, val_score.mean(axis=1), label='validation score')
    plt.legend()
    
    
```

```python
evaluation(model_1)
```

```python
evaluation(model_2)
```

```python
#Plot features based on the level of thier importances 

pd.DataFrame(model_1.feature_importances_, index=X_train.columns).plot.bar(figsize=(12, 8))
```
