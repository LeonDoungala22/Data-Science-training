# Step 3: Model optimisation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('Dataset/covid.csv', index_col=0, encoding = "ISO-8859-1")
data.head()
```

### PRE-PROCESSING

```python
df = data.copy()
df.head()
```

## Creation of sub-assemblies (following the step 1 and 2)

```python
missing_rate = df.isna().sum()/df.shape[0]
```

```python
blood_columns = list(df.columns[(missing_rate < 0.9) & (missing_rate >0.88)])
viral_columns = list(df.columns[(missing_rate < 0.80) & (missing_rate > 0.75)])
```

```python
key_columns = ['Patient age quantile', 'SARS-Cov-2 exam result']
```

```python
df = df[key_columns + blood_columns + viral_columns]
df.head()
```

## TrainTest - Cleanning - Encodage

```python
from sklearn.model_selection import train_test_split
```

```python
trainset, testset = train_test_split(df, test_size=0.2, random_state=0)
```

```python
trainset['SARS-Cov-2 exam result'].value_counts()
```

```python
testset['SARS-Cov-2 exam result'].value_counts()
```

```python
#Basic pre-prossesing functions

def encodage(df):
    code = {'negative':0,
            'positive':1,
            'not_detected':0,
            'detected':1}
    
    for col in df.select_dtypes('object').columns:
        df.loc[:,col] = df[col].map(code)
        
    return df

def feature_engineering(df):
    df['is sick'] = df[viral_columns].sum(axis=1) >= 1
    df = df.drop(viral_columns, axis=1)
    return df


def imputation(df):
    #df['is na'] = (df['Parainfluenza 3'].isna()) | (df['Leukocytes'].isna())
    #df = df.fillna(-999)
    df = df.dropna(axis=0)
    return  df
```

```python
#Basic preprossesing 

def preprocessing(df):
    
    df = encodage(df)
    df = feature_engineering(df)
    df = imputation(df)
    
    X = df.drop('SARS-Cov-2 exam result', axis=1)
    y = df['SARS-Cov-2 exam result']
    
    print(y.value_counts())
    
    return X, y
```

```python
X_train, y_train = preprocessing(trainset)
```

```python
X_test, y_test = preprocessing(testset)
```

# Evaluation procedure

```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import learning_curve
```

```python
def evaluation(model):
    
    model.fit(X_train, y_train)
    ypred = model.predict(X_test)
    
    print(confusion_matrix(y_test, ypred))
    print(classification_report(y_test, ypred))
    
    N, train_score, val_score = learning_curve(model, X_train, y_train,
                                              cv=4, scoring='f1',
                                               train_sizes=np.linspace(0.1, 1, 10))
    
    
    plt.figure(figsize=(12, 8))
    plt.plot(N, train_score.mean(axis=1), label='train score')
    plt.plot(N, val_score.mean(axis=1), label='validation score')
    plt.legend()
    
    
```

# Modelling

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
```

```python
preprocessor = make_pipeline(PolynomialFeatures(2, include_bias=False), SelectKBest(f_classif, k=10))
```

```python
#Models

RandomForest = make_pipeline(preprocessor, RandomForestClassifier(random_state=0))
AdaBoost = make_pipeline(preprocessor, AdaBoostClassifier(random_state=0))
SVM = make_pipeline(preprocessor, StandardScaler(), SVC(random_state=0,probability=True))
KNN = make_pipeline(preprocessor, StandardScaler(), KNeighborsClassifier())
```

```python
#Dict of models 

dict_of_models = {'RandomForest': RandomForest,
                  'AdaBoost' : AdaBoost,
                  'SVM': SVM,
                  'KNN': KNN
                 }
```

```python
for name, model in dict_of_models.items():
    print(name)
    evaluation(model)
```

## OPTIMIZATION

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
#I shoose to optimize SVM , it's generalize better than others 

#Print to know , the hyperparameters to optimize
SVM
```

```python
hyper_params = {'svc__gamma':[1e-3, 1e-4, 0.0005],
                'svc__C':[1, 10, 100, 1000, 3000],
                'pipeline__polynomialfeatures__degree':[2, 3],
                'pipeline__selectkbest__k': range(45, 60)
               }
```

```python
#Optimisation 

grid = RandomizedSearchCV(SVM, hyper_params, scoring='recall', cv=4,n_iter=40)      # Recall to avout false negatives cases 

#train de grid 
grid.fit(X_train, y_train)

print("- Best params: ",grid.best_params_)

y_pred = grid.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
evaluation(grid.best_estimator_)
```
