```python
#Dataset
boston_housing =tf.keras.datasets.boston_housing
(houses , targets),(_, _) = boston_housing.load_data()


#Rehsape dataset 
#images = images.reshape(-1,784)
#targets = targets.astype(float)

#Data Normalisation
#Scaler = StandardScaler()
#images = Scaler.fit(images) # Applicare il fit solo in training e il trasform su train e test
```

```python
houses.shape, targets.shape
```

```python
houses = pd.DataFrame(houses, columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'])
houses.head()
```

```python
targets = pd.DataFrame(targets , columns=['MEDV'])
targets.head()
```

```python
#Split dataset into train , test , val 

from sklearn.model_selection import train_test_split

images_train , images_test , targets_train , targets_test = train_test_split(houses, targets,test_size=0.2 , random_state=1)

#shapes 
print(images_train.shape , targets_train.shape)
print(images_test.shape , targets_test.shape)
```

# METOD 1 

## Model and add Layers

```python
# Flatten
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape = {13,}))

#Add Layers
model.add(tf.keras.layers.Dense(8, activation="relu", name='Input_layer')) 
model.add(tf.keras.layers.Dense(4, activation="relu", name='Hidden_layer_1'))
model.add(tf.keras.layers.Dense(1, activation=None, name='Output_layer')) 

model_output = model.predict(houses)
#print("- Image and target",images[0:1],targets[0:1])
print("- Image predicted:",model_output,targets)
```

### Model summary

```python
model.summary()
```

### Compile the model 

```python
# Compile the  model

model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer ="sgd",
    metrics=["accuracy"]
)
```

### Train the model 

```python
history = model.fit(images_train,targets_train,epochs=10 , validation_split=0.2)
```

### Plot validation curves 

```python
loss_curve = history.history["loss"]
acc_curve = history.history["accuracy"]

loss_val_curve = history.history["val_loss"]
acc_val_curve = history.history["val_accuracy"]
# plot loss
plt.figure(figsize=(10,5))
plt.plot(loss_curve, label="Train")
plt.plot(loss_val_curve, label="Val")
plt.legend(loc='upper left')
plt.title("Loss")
plt.xlabel('Epochs')
plt.ylabel('Values')
plt.legend()
plt.show()


#Plot accuracy
plt.figure(figsize=(10,5))
plt.plot(acc_curve, label="Train")
plt.plot(acc_val_curve, label="Val")
plt.legend(loc='upper left')
plt.title("Accuracy")
plt.xlabel('Epochs')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Get the loss and accuracy curve values for training and validation
loss_curve = history.history["loss"]
acc_curve = history.history["accuracy"]
loss_val_curve = history.history["val_loss"]
acc_val_curve = history.history["val_accuracy"]

# Create a figure with 2 subplots (1 row, 2 columns)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

# Add loss plot to first subplot (left)
ax1.plot(loss_curve, 'r-.', label="Train Loss")
ax1.plot(loss_val_curve, 'g--', label="Validation Loss")
ax1.legend(loc='upper right')
ax1.set_title("Loss", fontsize=20)
ax1.set_xlabel('Epochs', fontsize=15)
ax1.set_ylabel('Loss', fontsize=15)
ax1.grid(True)

# Add accuracy plot to second subplot (right)
ax2.plot(acc_curve, 'b-.', label="Train Accuracy")
ax2.plot(acc_val_curve, 'y--', label="Validation Accuracy")
ax2.legend(loc='lower right')
ax2.set_title("Accuracy", fontsize=20)
ax2.set_xlabel('Epochs', fontsize=15)
ax2.set_ylabel('Accuracy', fontsize=15)
ax2.grid(True)

# Display the figure
plt.show()
```

### Model evaluation 

```python
#Evaluate the model on train 

#Train Evaluate
loss_train , acc_train = model.evaluate(images_train,targets_train)
print("Loss_Train:",loss_train)
print("Accuracy_train:",acc_train)
```

```python

#Test  Evaluate

loss_test , acc_test  =model.evaluate(images_test,targets_test)
print("Loss_Test:",loss_test)
print("Accuracy_test:",acc_test)
```
