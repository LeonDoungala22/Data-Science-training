### Handle missing values

Possibles solutions : 

- Replacing a missing value via mean (if it is a numeric variable)
- Replace a missing value via mode (if it is a categorical variable)
- Replace a missing value by a default value, which is not admissible as a value for the column under consideration, e.g. "MISSING".
- Replace a missing value by rules based on the values of other columns
- Replace a missing column value by a statistical/ML-based algorithm, e.g. KNN

```python
#Count missing values by rows
count_na_rows()
```

```python
#Missig values and radio 
missing_values = Data.isnull().sum()/len(Data) # missing value ratio .  (Idealy near to 0 )
missing_values.sort_values(inplace=True)  # sorting

round(missing_values,2)
```

```python
# Now lets create a pandas dataframe of above result , to manipulate our dataset missing values values:
missing_values = (pd.DataFrame(missing_values))
missing_values.columns = ['count'] 
missing_values.index.names = ['Name']
round(missing_values,2)
```

```python
#Sorted missing values visualisation .
def plot_missing_values(missing_values):
    plt.figure(figsize=(20 , 10))
    sns.set(style="whitegrid", color_codes=True)
    sns.barplot(x = missing_values.index, y = 'count', data=missing_values)
    plt.xticks(rotation = 55)
    plt.show()
```

```python
plot_missing_values(missing_values)
```

```python
#Ho wmany total of missing values do we have ?
total_cells = np.product(Data.shape)
total_missing = missing_values.sum()

#Percent of missing values in dataset
percent_missing = (total_missing/total_cells)
print("Percent of missing values in dataset : ",percent_missing)
```

```python
#Rename some values
Data['GenderIDDesc'] = Data['GenderIDDesc'].map({'FEMALE': 'F', 'MALE': 'M'})

Data['AdvIssuedYNDesc'] = Data['AdvIssuedYNDesc'].fillna(Data['AdvIssuedYNDesc'].mode())
Data['ResultsIDDesc'] = Data['ResultsIDDesc'].fillna(Data['ResultsIDDesc'].mode())


Data.shape
```

```python
Data.head()
```

```python
#Handler some NaN values
Data['vaccination_yrs'].fillna(value=Data['vaccination_yrs'].mean(), inplace=True)
Data['GenderIDDesc'] = Data['GenderIDDesc'].fillna(Data['GenderIDDesc'].mode()[0]) #Replace NaN values by Gender mode 
Data['BreedIDDesc'] = Data['BreedIDDesc'].fillna(Data['BreedIDDesc'].mode()[0]) #Replace NaN values by BreedIDDesc mode 
Data['WhereBittenIDDesc'] = Data['WhereBittenIDDesc'].fillna(Data['WhereBittenIDDesc'].mode()[0]) #Replace NaN values by WhereBittenIDDesc mode 
Data['DispositionIDDesc'] = Data['DispositionIDDesc'].fillna(Data['DispositionIDDesc'].mode()[0]) #Replace NaN values by DispositionIDDesc mode 
Data['color'] = Data['color'].fillna(Data['color'].mode()[0]) #Replace NaN values by color's mode 
Data['victim_zip'] = Data['victim_zip'].fillna(Data['victim_zip'].mode()[0])
Data['vaccination_date'] = Data['vaccination_date'].fillna(Data['vaccination_date'].mode()[0])
Data['quarantine_date'] = Data['quarantine_date'].fillna(Data['quarantine_date'].mode()[0])
Data['release_date'] = Data['release_date'].fillna(Data['release_date'].mode()[0])
Data['head_sent_date'] = Data['head_sent_date'].fillna(Data['head_sent_date'].mode()[0])
Data['bite_date'] = Data['bite_date'].fillna(Data['bite_date'].mode()[0])
Data['SpeciesIDDesc'] = Data['SpeciesIDDesc'].fillna(Data['SpeciesIDDesc'].mode()[0])
   
Data.head(10)
```
